# Simple tests of UK RSP data access for VISTA

In this notebook we demonstrate some very simple test cases for accessing data from the VISTA VIRCAM processing including image access and catalogue queries.

```python
import lsst.daf.butler as dafButler
import lsst.afw.display as afwDisplay
import matplotlib.pyplot as plt
```

```python
butler = dafButler.Butler("/datasets/WP3.5/data/", writeable=False)
```

```python
for c in butler.registry.queryCollections():
    print(c)
```

```python
test = butler.get('deepCoadd_calexp',{'band':'K','tract':8524,'patch':48,'skymap':'hscPdr2'},collections='u/ir-shir1/DRP/videoMultiVisitFull')
```

```python
afwDisplay.setDefaultBackend('matplotlib') 
fig = plt.figure(figsize=(15,12))
afw_display = afwDisplay.Display(1)
afw_display.scale('asinh', 'zscale')
afw_display.mtv(test)
plt.gca().axis('off')
```
